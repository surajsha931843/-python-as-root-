```python
# there is also an alternative way to create a data i.e.
data = {'Hours':[1,2,3,4,5],'marks':[35,45,50,60,70]}# this is Dictionary method
df=pd.DataFrame(data)# now it is get converted in dataframe form
```

```python
# Visualize
plt.scatter(df['Hours'],df['marks'])
plt.xlabel('Hours studied')
plt.ylabel('Marks Scored')
plt.title('Hours vs Marks')
plt.show()
```

```python
# train the model
X= df[['Hours']]
Y= df['marks']
model = LinearRegression()
model.fit(X,Y)
```

```python
# predict
predicted_marks = model.predict([[6]])
print(predicted_marks)
```
